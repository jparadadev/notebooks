# Análisis Gauss Seidel

Análisis del cálculo del algoritmo de Gauss Seidel utilizando diferentes estrategias de paralelización.

### Imports de librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Gauss Seidel

```python
df_original = pd.read_csv('./datasets/gauss-seidel.1.1.csv')
df_p1 = pd.read_csv('./datasets/gauss-seidel-p1.1.128.csv')
df_p2 = pd.read_csv('./datasets/gauss-seidel-p2.1.128.csv')
df_p3 = pd.read_csv('./datasets/gauss-seidel-p3.1.128.csv')
df_p4 = pd.read_csv('./datasets/gauss-seidel-p4.1.128.csv')

df_p1.head()
```

```python
df_p1 = df_p1.groupby(['threads'], as_index=True)['time'].median().reset_index(name='time')
df_p2 = df_p2.groupby(['threads'], as_index=True)['time'].median().reset_index(name='time')
df_p3 = df_p3.groupby(['threads'], as_index=True)['time'].median().reset_index(name='time')
df_p4 = df_p4.groupby(['threads'], as_index=True)['time'].median().reset_index(name='time')

df_p1.head()
```

```python
df_p1.plot(x='threads', y='time', title='Gauss Seidel P1')
df_p2.plot(x='threads', y='time', title='Gauss Seidel P2')
df_p3.plot(x='threads', y='time', title='Gauss Seidel P3')
df_p4.plot(x='threads', y='time', title='Gauss Seidel P4')

plt.show()
```
