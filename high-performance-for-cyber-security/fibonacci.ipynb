# Análisis Fibonacci

Análisis del cálculo del algoritmo recursivo Fibonacci utilizando diferentes técnicas de paralelización.

### Imports de librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Fibonacci original

```python
df_10 = pd.read_csv('./datasets/fibonacci.10.1.10.csv')
df_20 = pd.read_csv('./datasets/fibonacci.20.1.10.csv')
df_30 = pd.read_csv('./datasets/fibonacci.30.1.10.csv')

df_10.head()
```

```python
df_10_med = df_10.groupby(['threads'], as_index=True)['time'].median().reset_index(name='time')
df_20_med = df_20.groupby(['threads'], as_index=True)['time'].median().reset_index(name='time')
df_30_med = df_30.groupby(['threads'], as_index=True)['time'].median().reset_index(name='time')

df_10_med.head()
```

```python
df_10_med.plot(x='threads', y='time', title='Fibonacci - 10')
df_20_med.plot(x='threads', y='time', title='Fibonacci - 20')
df_30_med.plot(x='threads', y='time', title='Fibonacci - 30')

plt.show()
```
